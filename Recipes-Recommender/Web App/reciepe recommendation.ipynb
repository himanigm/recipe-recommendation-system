```python
# Convert string representation of list to actual list
def clean_ingredients(ingredient_str):
    try:
        ingredients = ast.literal_eval(ingredient_str)
        return ', '.join([i.strip().lower() for i in ingredients])
    except:
        return ''

recipes_df['clean_ingredients'] = recipes_df['ingredients'].apply(clean_ingredients)

# Drop rows where ingredients are missing
recipes_df = recipes_df[recipes_df['clean_ingredients'] != ''].reset_index(drop=True)
```

```python
def recommend_recipes(user_ingredients, recipes_df, top_n=5):
    vectorizer = CountVectorizer()
    ingredient_matrix = vectorizer.fit_transform(recipes_df['clean_ingredients'])

    input_str = ', '.join([ing.strip().lower() for ing in user_ingredients])
    input_vector = vectorizer.transform([input_str])

    cosine_sim = cosine_similarity(input_vector, ingredient_matrix)
    top_indices = cosine_sim[0].argsort()[::-1][:top_n]

    return recipes_df.iloc[top_indices][['name', 'clean_ingredients']]
```

```python
# 🧑 User Inputs Ingredients
user_input_ingredients = ['egg', 'rice', 'onion']

# 🔍 Get top 5 recipe recommendations
recommended = recommend_recipes(user_input_ingredients, recipes_df, top_n=5)

# 📋 Display results
print("\nRecommended Recipes:")
print(recommended)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
recipes_df = pd.read_csv(r'C:\Users\Rohit94\Documents\Recipes-Recommender\Web App\datasets/rr-recipes.csv')

# Clean ingredients
def clean_ingredients(ingredient_str):
    if isinstance(ingredient_str, str):
        return ', '.join([i.strip().lower() for i in ingredient_str.split(',')])
    return ''

recipes_df['clean_ingredients'] = recipes_df['ingredients'].apply(clean_ingredients)
recipes_df = recipes_df[recipes_df['clean_ingredients'].str.strip() != ''].reset_index(drop=True)

# Parse user input string like "chocolate,cheese"
def parse_ingredient_input(input_string):
    return [item.strip().lower() for item in input_string.split(',') if item.strip()]

# Recommend recipes
def recommend_recipes(user_ingredients, recipes_df, top_n=5):
    vectorizer = CountVectorizer()
    ingredient_matrix = vectorizer.fit_transform(recipes_df['clean_ingredients'])

    input_str = ', '.join(user_ingredients)
    input_vector = vectorizer.transform([input_str])

    cosine_sim = cosine_similarity(input_vector, ingredient_matrix)
    top_indices = cosine_sim[0].argsort()[::-1][:top_n]

    return recipes_df.iloc[top_indices][[
        'photo_url',
        'title',
        'ready_time',
        'prep_time',
        'cook_time',
        'directions',
        'url'
    ]]

# 🍫🍰 Example input
user_input_string = "Rice"
user_ingredients = parse_ingredient_input(user_input_string)

# Get recommendations
recommended = recommend_recipes(user_ingredients, recipes_df)

# Display
pd.set_option('display.max_colwidth', None)
display(recommended.rename(columns={
    'photo_url': 'Image',
    'title': 'Recipe Name',
    'ready_time': 'Ready In (min)',
    'prep_time': 'Prep Time (min)',
    'cook_time': 'Cook Time (min)',
    'directions': 'Recipe Procedure',
    'url': 'Source URL'
}))

```

```python
# 📦 Import required libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 📥 Load dataset
recipes_df = pd.read_csv(r'C:\Users\Rohit94\Documents\Recipes-Recommender\Web App\datasets/rr-recipes.csv')

# ✅ Clean the ingredients (standardize case and spacing)
def clean_ingredients(ingredient_str):
    if isinstance(ingredient_str, str):
        return ', '.join([i.strip().lower() for i in ingredient_str.split(',')])
    return ''

recipes_df['clean_ingredients'] = recipes_df['ingredients'].apply(clean_ingredients)
recipes_df = recipes_df[recipes_df['clean_ingredients'].str.strip() != ''].reset_index(drop=True)

# ✅ Parse user input string like "chocolate, cheese"
def parse_ingredient_input(input_string):
    return [item.strip().lower() for item in input_string.split(',') if item.strip()]
def recommend_recipes_tfidf(user_ingredients, recipes_df, top_n=5):
    vectorizer = TfidfVectorizer()
    ingredient_matrix = vectorizer.fit_transform(recipes_df['clean_ingredients'])

    input_str = ', '.join(user_ingredients)
    input_vector = vectorizer.transform([input_str])

    cosine_sim = cosine_similarity(input_vector, ingredient_matrix)
    top_indices = cosine_sim[0].argsort()[::-1][:top_n]

    return recipes_df.iloc[top_indices][[
        'photo_url', 'title', 'ready_time', 'prep_time',
        'cook_time', 'directions', 'url'
    ]]
def jaccard_similarity(set1, set2):
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union) if union else 0

def recommend_recipes_jaccard(user_ingredients, recipes_df, top_n=5):
    user_set = set(user_ingredients)
    recipes_df['jaccard_score'] = recipes_df['clean_ingredients'].apply(
        lambda x: jaccard_similarity(user_set, set(x.split(',')))
    )
    top_matches = recipes_df.sort_values('jaccard_score', ascending=False).head(top_n)
    return top_matches[[
        'photo_url', 'title', 'ready_time', 'prep_time',
        'cook_time', 'directions', 'url'
    ]]
# 🍫 Example Input
user_input_string = "chocolate, cheese"
user_ingredients = parse_ingredient_input(user_input_string)

# 💡 Choose one of the models:
# recommended = recommend_recipes_tfidf(user_ingredients, recipes_df)
recommended = recommend_recipes_jaccard(user_ingredients, recipes_df)

# 📋 Show results
pd.set_option('display.max_colwidth', None)
display(recommended.rename(columns={
    'photo_url': 'Image',
    'title': 'Recipe Name',
    'ready_time': 'Ready In (min)',
    'prep_time': 'Prep Time (min)',
    'cook_time': 'Cook Time (min)',
    'directions': 'Recipe Procedure',
    'url': 'Source URL'
}))
```
